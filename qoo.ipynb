```python
def test(a):
    boxes=a.reshape((3,3,-1)).transpose((0,2,1)).reshape((3,3,3,3))
    columns = boxes.sum(axis=3).sum(axis=0).ravel()
    rows = boxes.sum(axis=1).sum(axis=1).ravel()
    sboxes = boxes.sum(axis=3).sum(axis=2)
    test1 = (rows==45).all()
    test2 = (columns==45).all()
    test3 = (sboxes==45).all()
    return test1 and test2 and test3
```
