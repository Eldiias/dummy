# Comparison with sklearn PCA

```python
from sklearn.decomposition import PCA
```

```python
pca1=PCA(2) # you can specify explained variance ratio instead of number of components
```

```python
pca1.fit(X)
```

```python
pca1.explained_variance_ratio_ # we can see the difference. But it is simply because the data is not standardized
```

```python
pca1.transform(X) #therefore data is different
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X1 = pd.DataFrame(scaler.transform(X),columns=X.columns.values.tolist())
```

```python
pca1=PCA(2)
pca1.fit(X1)
```

```python
pca1.explained_variance_ratio_.sum() # after standartization we obtain 100% the same result
```

```python
pca1.transform(X) # here as well
```

```python
pca_inv=pca1.inverse_transform(pca1.transform(X1)) # inverse transformation of pca will lead to standardized values
```

```python
pd.DataFrame(scaler.inverse_transform(pca_inv), columns=X.columns) #Dataframe of inverse transformation of scaling
```

# Conclusion

Statsmodels PCA implementation has a few advantages:
1. Data is standardized by itself
2. No need of fitting data. Everything is calculated on the initialisation step
3. Inverse calculations are also omitted - you simply call invere transformation function and it gives you inverse transformation both for PCA and for standartisation.
4. Returns you pandas DataFrame object keeping all the names.

But at the same time sklearn implementation has following pros:
1. You can specify the explained variance ratio instead of number of components
2. If you already standardized your data, you don't need to specify that (we didn't see it here, but in statsmodels you need to define that you already have standardized your data)
3. Structure follows sklearn structure that simplifies the process (you know that first you create the object, then you fit it and so on)

In terms of scientifical importance, statsmodels approach is much more powerfull, since it proposes you different approaches of PC calculation (both analytical and numerical), while sklearn one is optimized for production.
